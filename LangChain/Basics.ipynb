```python
from Models.groq import llm
from langchain.agents import create_agent
from langchain.tools import tool
system_prompt = """
You are an execution agent.

Behavior rules:
- You MUST use a tool if one is relevant.
- For multiplication requests, you MUST call the tool `find_value`.
- NEVER respond with plain text for calculations.
- Extract numbers from the user message.
- Call the tool with arguments `a` and `b`.
- Your final answer MUST come from the tool output.
- If you do not call the tool, the response is invalid.
"""

@tool
def find_value(a,b):
    """
    Multiply a and b and return 
    """
    print("tool invoked")
    return a*b
agent=create_agent(llm,tools=[find_value],system_prompt=system_prompt)

for stream in agent.stream({"message":"find the value of 4 and 5"},stream_mode="values"):
    print(stream["messages"][-1].pretty_print())
```
