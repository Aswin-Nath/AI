```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
read_path=r"C:\Users\aswin\.cache\kagglehub\datasets\terryeppler\regulations-pdf\versions\1\Statements of Federal Federal Financial Accounting Concepts and Standards.pdf"
```

```python
def load_doc(path):
    return PyPDFLoader(path).load()
```

```python
doc=load_doc(read_path)
```

```python
sffac_pages=[]
sffas_pages=[]
interpretation_pages=[]
technical_bulletin_checker="Technical Bulletin ".lower()
technical_bulletin_pages=[]
technical_releases_checker="Technical Releases ".lower()
technical_releases_pages=[]
staff_implementation_checker="Staff Implementation Guidance ".lower()
staff_implementation_pages=[]
sffas_checker="Statement of Federal Financial Accounting Standards".lower()
sffac_checker="Statement of Federal Financial Accounting Concepts".lower()
interpretation_checker="Interpretation of Federal Financial Accounting Standards".lower()
def get_metadata(metadata):
    wanted_keys=['total_pages','page_label','page','title','source']
    return {key:metadata[key] for key in wanted_keys}

for page in doc:
    metadata=page.metadata
    wanted_metadata=get_metadata(metadata)
    individual_page=page.page_content
    individual_page=individual_page.lower()
    needed_data={"page_content":individual_page,"metadata":wanted_metadata}
    if sffas_checker in individual_page:
        sffas_pages.append(needed_data)
    if sffac_checker in individual_page:
        sffac_pages.append(needed_data)
    if interpretation_checker in individual_page:
        interpretation_pages.append(needed_data)
    if technical_bulletin_checker in individual_page:
        technical_bulletin_pages.append(needed_data)
    if technical_releases_checker in individual_page:
        technical_releases_pages.append(needed_data)
    if staff_implementation_checker in individual_page:
        staff_implementation_pages.append(needed_data)
```

```python
tot=0
for i in [sffac_pages,sffas_pages,interpretation_pages,technical_bulletin_pages,technical_releases_pages,staff_implementation_pages]:
    tot+=len(i)
    print(len(i))
print(tot,len(doc))
```
