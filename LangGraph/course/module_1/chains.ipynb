```python

def multiply(x:int,y:int)->int:
    """
    return the multiplication of the number x with y
    Args:
        x:int
        y:int
        
    """
    return x*y
```

```python
tools=[multiply]
```

```python
llm_with_tools=llm.bind_tools(tools)
```

```python
tool_call=llm_with_tools.invoke([HumanMessage(content=f"I want to know the value when 88 is multiplied with 100",name="Aswinnath")])
```

```python
tool_call.pretty_print
```

```python
tool_calls=tool_call.tool_calls
```
